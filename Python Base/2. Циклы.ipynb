ПЕРЕБОРНЫЙ ЦИКЛ **`for`** 

> **Цикл** — конструкция языка, позволяющая многократно повторять заданный набор инструкций.



```
for имя переменная in набор данных:
	действия для каждого элемента из набора (тело цикла)
```

```
for i in 2, 12, 3, 1, 50:
	print(i ** 2) 
```

```
for char in "eHllo, world!":
	print(char)           
```

Функция **`range()`**

Позволяет генерировать последовательный набор чисел по заданному началу, концу и шагу.

Функция может принять до трёх параметров:

- `range(end)` создаст набор чисел от `0` до `end - 1` с шагом `1`;
- `range(start, end)` создаст набор чисел от `start` до `end - 1` с шагом `1`;
- `range(start, end, step)` создаст набор чисел от `start` до `end - 1` с шагом `step`.

Примеры:

- `range(5)` сгенерирует числа от 0 до 4;
- `range(3, 7)` сгенерирует числа от 3 до 6;
- `range(2, 9, 3)` сгенерирует числа от 2 до 9 с шагом 3 (это 2, 5 и 8);

```
start = 2
end = 9
step = 2

print(*range(end))              
print(range(start, end))       
print(list(range(start, end, step))) 
```

```
var = 0
for i in range(9): 
	if i < 6:
		var += i*2 # 2, 6, 12, 20, 30
	else:
		var += i # 36, 43, 51
print(var)
```

```
count = int(input('Количество чисел: ')) 
s = 0

for i in range(count):                   
  print('Введите число №', i, end=': ') 
  number = int(input())                     
  s += number                               

print('Сумма всех чисел:', s) 
```

ЦИКЛ С ПРЕДУСЛОВИЕМ **`while`**

```
while условие:
	действия, выполняющиеся при истинности условия (тело цикла)
```

```
n = 0

while n < 10:
  while n % 2 == 0:
    print(n)
    n += 3
  n += 3
```



> Все переменные, которые используются в условии цикла, должны быть инициализированы **ДО** начала цикла.



```
s = 0                         

cur = -1                  
															
while cur != 0:               
	cur = int(input())          
	s += cur                    

print('Сумма всех чисел:', s)
```

```
s = 0
number = int(input())
s += number                
															
while s < 100:               
	number = int(input())          
	s += number                   

print('Сумма всех чисел:', s)
```

```
n = int(input())
s = 0
while n != 0:
  number = int(input())
  s += number
  n -= 1
print(s)
```

Изменение хода выполнения цикла

Оператор **`break`**

```
s = 0                

while True:          
	cur = int(input()) 
	if cur == 0:       
		break            
	s += cur           
                     
print('Сумма введённых чисел:', s)
```

```
s = 0
for i in range(9):
  cur = int(input())
  if cur % 2 == 1:
    break
  s += cur
print(s) 
```

Оператор **`continue`** позволяет завершить только текущую итерацию цикла, а не цикл полностью.

```
count = int(input('Количество чисел: '))
s = 0

for i in range(count): 
	cur = int(input())   
	if cur % 2 == 1:     
		continue           
	s += cur             

print('Сумма чётных чисел:', s)
```

```
for i in range(9):
  if i >= 0:
    continue
  print(i)
```

Вложенные циклы

```
n = int(input('Введите размер таблицы: '))

for i in range(1, n + 1):   
	for j in range(1, n + 1): 
		print(i * j, end='\t')  
	print() 
```
